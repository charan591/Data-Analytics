# Bed Vacancy Details Under Covid Health Centres And
# Covid District Hospitals In amil_nadu

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv("bed_vacancy_details.csv")
df
```

```python
df_cleaned = df.rename(columns={
    "Name of the District": "District",
    "Earmarked beds for COVID with O2 supply under CHC and CDH as on 15.06.2022": "Total_O2_Beds",
    "Earmarked beds for COVID with Non O2 supply under CHC and CDH as on 15.06.2022": "Total_NonO2_Beds",
    "Earmarked beds for COVID in ICU under CHC and CDH as on 15.06.2022": "Total_ICU_Beds",
    "Occupancy of beds with O2 supply under CHC and CDH as on 15.06.2022": "Occupied_O2_Beds",
    "Occupancy of beds with Non O2 supply under CHC and CDH as on 15.06.2022": "Occupied_NonO2_Beds",
    "Occupancy of beds in ICU under CHC and CDH as on 15.06.2022": "Occupied_ICU_Beds",
    "Vacancy of beds with O2 supply under CHC and CDH as on 15.06.2022": "Vacant_O2_Beds",
    "Vacancy of beds with Non O2 supply under CHC and CDH as on 15.06.2022": "Vacant_NonO2_Beds",
    "Vacancy of beds in ICU under CHC and CDH as on 15.06.2022": "Vacant_ICU_Beds",
    "Total Vacant beds under CHC and CDH as on 15.06.2022": "Total_Vacant_Beds"
})
df
```

```python
df_cleaned = df_cleaned.drop(columns=["S.No"], errors='ignore')
df
```

```python
total_vacant_beds = df_cleaned["Total_Vacant_Beds"].sum()
```

```python
total_beds = df_cleaned["Total_O2_Beds"] + df_cleaned["Total_NonO2_Beds"] + df_cleaned["Total_ICU_Beds"]
total_occupied_beds = df_cleaned["Occupied_O2_Beds"] + df_cleaned["Occupied_NonO2_Beds"] + df_cleaned["Occupied_ICU_Beds"]
overall_occupancy_rate = (total_occupied_beds.sum() / total_beds.sum()) * 100
df
```

```python
highest_vacancy_district = df_cleaned.loc[df_cleaned["Total_Vacant_Beds"].idxmax(), ["District", "Total_Vacant_Beds"]]
lowest_vacancy_district = df_cleaned.loc[df_cleaned["Total_Vacant_Beds"].idxmin(), ["District", "Total_Vacant_Beds"]]
df
```

```python
summary_report = {
    "Total Vacant Beds": total_vacant_beds,
    "District with Highest Vacancy": highest_vacancy_district.to_dict(),
    "District with Lowest Vacancy": lowest_vacancy_district.to_dict(),
    "Overall Bed Occupancy Rate (%)": round(overall_occupancy_rate, 2)
}

print(summary_report)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="District", y="Total_Vacant_Beds", data=df_cleaned, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("District")
plt.ylabel("Total Vacant Beds")
plt.title("Vacant Beds per District")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df_cleaned.select_dtypes(include=[np.number]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Bed Occupancy Data")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
bed_types = [total_occupied_beds.sum(), total_vacant_beds]
labels = ["Occupied Beds", "Vacant Beds"]
plt.pie(bed_types, labels=labels, autopct="%1.1f%%", colors=["red", "green"])
plt.title("Overall Bed Occupancy Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="District", y="Total_Vacant_Beds", data=df_cleaned, marker="o", color="blue")
plt.xticks(rotation=90)
plt.xlabel("District")
plt.ylabel("Total Vacant Beds")
plt.title("Trend of Vacant Beds per District")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned[["Vacant_O2_Beds", "Vacant_NonO2_Beds", "Vacant_ICU_Beds"]], palette="Set2")
plt.xlabel("Bed Type")
plt.ylabel("Vacant Beds")
plt.title("Distribution of Vacant Beds by Type")
plt.show()
```
